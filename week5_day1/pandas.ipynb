```python
import pandas as pd

# Create initial DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
}
df = pd.DataFrame(data)

#new column
df['Age_in_5yrs'] = df['Age'] + 5
print("After Adding Column:\n", df)

#Drop column
df = df.drop('Age_in_5yrs', axis=1)
print("\nAfter Removing Column:\n", df)
```

```python
# Create a new DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob'],
    'Score': [85, 90, 95, 80, 70]
}
df = pd.DataFrame(data)

#  Group by Name and calculate mean score
grouped = df.groupby('Name').mean(numeric_only=True)
print("\nGrouped & Mean Score:\n", grouped)
```
